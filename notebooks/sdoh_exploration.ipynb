## Step 1: Load the Data

```python
import pandas as pd
import numpy as np

# Load the cleaned SDOH county dataset (pre-processed from AHRQ source)
df = pd.read_csv('data/sdoh_county_2023.csv', dtype={'county_fips': str})

print("✅ Dataset loaded successfully!")
print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
```

## Step 2: Get to Know the Structure

```python
# Preview first 5 rows
df.head()
```

```python
# Dimensions
print(f"Rows (counties): {df.shape[0]:,}")
print(f"Columns:         {df.shape[1]}")
```

```python
# Column names, data types, and non-null counts
df.info()
```

## Step 3: Explore Individual Columns

```python
# Summary statistics for all numerical columns
df.describe().round(2)
```

```python
# Counties per state (top 10)
print("Counties per state (top 10):")
print(df['state'].value_counts().head(10))
```

```python
# Counties by Census region
print("Counties by Census region:")
print(df['region'].value_counts())
```

```python
# Poverty rate distribution
print("Poverty rate distribution:")
print(df['poverty_pct'].describe().round(2))
print()
print(f"Counties with poverty > 20%: {(df['poverty_pct'] > 20).sum():,}")
print(f"Counties with poverty > 30%: {(df['poverty_pct'] > 30).sum():,}")
```

## Step 4: Check for Missing Values & Duplicates

```python
# Missing value report
missing = df.isna().sum()
missing_pct = (df.isna().mean() * 100).round(2)

missing_report = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
}).sort_values('Missing Count', ascending=False)

print("Missing Value Report:")
has_missing = missing_report[missing_report['Missing Count'] > 0]

if len(has_missing) > 0:
    print(has_missing.to_string())
else:
    print("✅ No missing values found!")
```

```python
# Check for duplicate county FIPS codes
dupes = df.duplicated(subset=['county_fips']).sum()
print(f"Duplicate county FIPS codes: {dupes}")
print(f"Fully duplicate rows:        {df.duplicated().sum()}")

if dupes == 0:
    print("✅ All counties are unique!")
```

## Step 5: Filter, Sort & Explore Patterns

```python
# .loc[] — Filter counties where poverty rate is above 25%
high_poverty = df.loc[df['poverty_pct'] > 25].copy()

print(f"High-poverty counties (poverty > 25%): {len(high_poverty):,}")
print()
print(high_poverty[['county', 'state', 'poverty_pct', 'uninsured_pct', 'median_income']]
      .sort_values('poverty_pct', ascending=False)
      .head(15)
      .to_string(index=False))
```

```python
# sort_values() — Top 15 counties by uninsured rate
print("Top 15 counties by uninsured rate:")
print(df[['county', 'state', 'uninsured_pct', 'poverty_pct', 'median_income']]
      .sort_values('uninsured_pct', ascending=False)
      .head(15)
      .to_string(index=False))
```

```python
# iloc[] — select rows and columns by position
print("First 5 rows, first 6 columns (using .iloc[]):")
print(df.iloc[0:5, 0:6])
```

```python
# .loc[] — Average broadband access by Census region
print("Average broadband access by Census region:")
for region in df['region'].dropna().unique():
    avg = df.loc[df['region'] == region, 'broadband_pct'].mean()
    print(f"  {region:<15} {avg:.1f}% of households with broadband")
```

```python
# Boolean indexing — healthcare deserts (> 30 miles from emergency dept)
healthcare_deserts = df.loc[df['dist_emergency_dept'] > 30].copy()

print(f"Counties with median ER distance > 30 miles: {len(healthcare_deserts):,}")
print()
print(healthcare_deserts[['county', 'state', 'dist_emergency_dept',
                           'dist_trauma_center', 'poverty_pct']]
      .sort_values('dist_emergency_dept', ascending=False)
      .head(10)
      .to_string(index=False))
```

## Step 6: Explore Racial Disparities in Poverty

```python
# Compare average poverty rates by racial group across all counties
print("Average poverty rate by racial group (across all U.S. counties):")
print(f"  Black population:    {df['poverty_pct_black'].mean():.1f}%")
print(f"  Hispanic population: {df['poverty_pct_hispanic'].mean():.1f}%")
print(f"  White population:    {df['poverty_pct_white'].mean():.1f}%")
print(f"  Overall avg:         {df['poverty_pct'].mean():.1f}%")
```

## Step 7: Social Vulnerability Score

```python
# Composite vulnerability score — already computed in the dataset
# Formula: (poverty + uninsured + unemployment + (100-broadband) + (100-education)) / 5
# Higher score = more vulnerable county

print("Most Vulnerable Counties (Top 15):")
print(df[['county', 'state', 'region', 'vulnerability_score',
          'poverty_pct', 'uninsured_pct', 'unemployment_pct']]
      .sort_values('vulnerability_score', ascending=False)
      .head(15)
      .round(1)
      .to_string(index=False))
```

```python
# Score distribution
print("Vulnerability Score Summary:")
print(df['vulnerability_score'].describe().round(2))
print()
print(f"High vulnerability  (score > 35): {(df['vulnerability_score'] > 35).sum():,} counties")
print(f"Medium vulnerability (20–35):     {((df['vulnerability_score'] >= 20) & (df['vulnerability_score'] <= 35)).sum():,} counties")
print(f"Low vulnerability   (score < 20): {(df['vulnerability_score'] < 20).sum():,} counties")
```

```python
# Least vulnerable counties
print("Least Vulnerable Counties (Top 10):")
print(df[['county', 'state', 'region', 'vulnerability_score', 'poverty_pct', 'median_income']]
      .sort_values('vulnerability_score', ascending=True)
      .head(10)
      .round(1)
      .to_string(index=False))
```
